```python
#%matplotlib inline

%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D          #for making 3d Axes (canvas)
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
def animate_3d(i):
    Z = 0.2*np.sin(2 * np.pi * (X + Y - 0.01 * i))
    surface_ax.clear() #very important, try running without it
    surface = surface_ax.plot_wireframe(X, Y, Z, rstride=1, cstride=0)
    return (surface)
```

```python
surface_fig = plt.figure()
surface_ax = surface_fig.add_subplot(111, projection = '3d')

x = np.linspace(0, 2, 20)
y = np.linspace(0, 2, 10)
X, Y = np.meshgrid(x,y)
Z = 0.2*np.sin(2 * np.pi * (X + Y - 0.01 ))
surface = surface_ax.plot_wireframe(X, Y, Z, rstride=1, cstride=0, color = 'b')

#ax.set_xlim3d([0.0, 1.0])
surface_ax.set_xlabel('X')
#ax.set_ylim3d([0.0, 1.0])
surface_ax.set_ylabel('Y')
surface_ax.set_zlim3d([-0.5, 0.5])
surface_ax.set_zlabel('Z')
surface_ax.set_title('3D Test')
```

```python
surface_ani = animation.FuncAnimation(surface_fig, animate_3d, frames=200, interval=20, blit=False)
```

```python
HTML(surface_ani.to_html5_video())
```
