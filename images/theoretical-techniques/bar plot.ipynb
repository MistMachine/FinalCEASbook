```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
z = 0
xs = np.linspace(1,20,20)
ys = np.ones_like(xs)
cs = ['r'] * len(xs)

ax.bar(xs, ys, 0, zdir='x', color=cs, alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plt.show()
```
