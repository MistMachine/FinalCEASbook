```python
%matplotlib inline

#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D          #for making 3d Axes (canvas)
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
def init():
    line.set_data([], [])
    line.set_3d_properties([])
    return (line,)

def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.zeros_like(x)+0.5
    z = 0.2*np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    line.set_3d_properties(z)
    return (line,)
```

```python
x = np.linspace(0, 2, 1000)
y = np.zeros_like(x)+0.5
z = 0.2*np.sin(2 * np.pi * (x - 0.01))

fig = plt.figure()
ax = fig.add_subplot(111, projection= '3d')
line, = ax.plot(x, y, z, lw=2)
#ax.set_xlim3d([0.0, 1.0])
ax.set_xlabel('X')
#ax.set_ylim3d([0.0, 1.0])
ax.set_ylabel('Y')
#ax.set_zlim3d([0.0, 1.0])
ax.set_zlabel('Z')
ax.set_title('3D Test')
plt.show()
```

```python
line_ani = animation.FuncAnimation(fig, animate, frames=100, interval=20, blit=True)
```
